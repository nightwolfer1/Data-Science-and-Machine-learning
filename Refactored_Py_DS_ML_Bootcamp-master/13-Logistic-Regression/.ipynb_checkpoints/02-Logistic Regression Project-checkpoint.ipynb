# Logistic Regression Project 

In this project we will be working with a fake advertising data set, indicating whether or not a particular internet user clicked on an Advertisement. We will try to create a model that will predict whether or not they will click on an ad based off the features of that user.

This data set contains the following features:

* 'Daily Time Spent on Site': consumer time on site in minutes
* 'Age': cutomer age in years
* 'Area Income': Avg. Income of geographical area of consumer
* 'Daily Internet Usage': Avg. minutes a day consumer is on the internet
* 'Ad Topic Line': Headline of the advertisement
* 'City': City of consumer
* 'Male': Whether or not consumer was male
* 'Country': Country of consumer
* 'Timestamp': Time at which consumer clicked on Ad or closed window
* 'Clicked on Ad': 0 or 1 indicated clicking on Ad

# Logistic Regression

Now it's time to do a train test split, and train our model!

You'll have the freedom here to choose columns that you want to train on!

** Split the data into training set and testing set using train_test_split**

```python
from sklearn.model_selection import train_test_split
#ill start using data from all numerical columns to train and see how well the model does.
#train and test data generated below
X_train,X_test,y_train,y_test=train_test_split(Copy_data.drop('Clicked on Ad',axis=1),Copy_data['Clicked on Ad'],test_size=0.30,random_state=101)
```

** Train and fit a logistic regression model on the training set.**

```python
#now lets train the logisticRegression model, first we import the object
from sklearn.linear_model import LogisticRegression
#initialize the mode
Logmodel=LogisticRegression()
#Time to train model
Logmodel.fit(X_train,y_train)
```

## Predictions and Evaluations
** Now predict values for the testing data.**

```python
#now we predict using the test data on the model
predictions=Logmodel.predict(X_test)
```

** Create a classification report for the model.**

```python
from sklearn.metrics import confusion_matrix
#Lets look at the confusion matrix.
print(confusion_matrix(y_test,predictions))

#Precision = True positive / all positive 
prec=129/(129+8) # for positive outcome (1)
# Recall = True positive / True positive+ False negative , all predicted correct values devided by the relevant instances  
recall=149/(149+14)# for positive outcome (1)
#F1 score is is the harmonic mean of precision and recall, measure of test accuracy 
2*prec*recall/(prec+recall)
```

```python
from sklearn.metrics import classification_report
#Classification report for model
print(classification_report(y_test,predictions))

#model has a good precision recall and f1score above 90 % for each which i consider good with the data available.
```
